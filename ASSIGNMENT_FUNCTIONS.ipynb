Q1. What is the difference between a function and a method in Python?

ANS. The main difference between a function and a method in Python is how they are called and where they belong.

Function

A function is a block of reusable code that is defined using def and can exist independently.

It is called directly using its name.

It can take arguments and return values.

Functions can be built-in (e.g., print(), len()) or user-defined.

Example of a function:

def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))  # Output: Hello, Alice!


Method

A method is a function that is associated with an object and is called on that object.

It is defined inside a class and takes self as the first parameter (in instance methods).

Methods are used to operate on the attributes of an instance or class.

Example of a method:

class Person:
    def _init_(self, name):
        self.name = name
    
    def greet(self):
        return f"Hello, {self.name}!"

p = Person("Alice")
print(p.greet())  # Output: Hello, Alice!




Q2. Explain the concept of function arguments and parameters in Python?

Ans. In Python, function arguments and parameters are used to pass data into functions. While they are often used interchangeably, they have distinct meanings:

1. Parameters vs. Arguments

Parameters are the variables listed in a function definition.

Arguments are the actual values passed to the function when calling it.


Example:

def greet(name):  # 'name' is a parameter
    print(f"Hello, {name}!")

greet("Alice")  # "Alice" is an argument

2. Types of Function Arguments in Python

Python supports several types of arguments:

(i) Positional Arguments

These are passed in the same order as the function’s parameters.

The number of arguments must match the number of parameters.


Example:

def add(a, b):
    return a + b

print(add(5, 3))  # Output: 8

(ii) Keyword Arguments

These are passed using parameter names.

The order does not matter.


Example:

def introduce(name, age):
    print(f"My name is {name} and I am {age} years old.")

introduce(age=25, name="Alice")  # Output: My name is Alice and I am 25 years old.


---

(iii) Default Arguments

Parameters can have default values, making them optional when calling the function.


Example:

def greet(name="Guest"):
    print(f"Hello, {name}!")

greet()        # Output: Hello, Guest!
greet("Alice") # Output: Hello, Alice!


---

(iv) Variable-Length Arguments (*args)

Allows passing multiple positional arguments as a tuple.


Example:

def sum_all(*args):
    return sum(args)

print(sum_all(1, 2, 3, 4))  # Output: 10


---

(v) Keyword Variable-Length Arguments (**kwargs)

Allows passing multiple keyword arguments as a dictionary.


Example:

def display_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

display_info(name="Alice", age=25, city="New York")
# Output:
# name: Alice
# age: 25
# city: New York

Q4. What is the purpose of the return statement in a Python function?

Ans. The return statement in a Python function serves the purpose of sending a value back to the caller and terminating the function execution.

1. Returning a Value

A function can return a value using return, allowing the caller to store and use the result.

Example:

def add(a, b):

    return a + b  # Returns the sum

---

2. Returning Multiple Values

Python allows returning multiple values as a tuple.

Example:

    return 10, 20  # Returns a tuple

3. Returning Early (Terminating a Function)

Once return is encountered, the function exits immediately.

        return "Even"

    return "Odd"  # This runs only if n is not even

print(check_even(4))  # Output: Even

4. Returning Nothing (None)

If no return statement is provided, or if return is used without a value, the function returns None by default.

return is essential in recursive functions to pass computed values up the recursion stack.

    if n == 1:

        return 1

    return n * factorial(n - 1)

You can use return inside a loop to exit the function early.

Q5. What are iterators in Python and how do they differ from iterables?

Ans. In Python, iterators and iterables are both used for looping, but they are different concepts.


---

1. What is an Iterable?

An iterable is any object that can return an iterator. It contains data that can be looped over (like lists, tuples, dictionaries, strings, etc.).

It has the special method _iter_(), which returns an iterator.

Examples: lists, tuples, dictionaries, strings, sets, etc.


Example of an iterable:

numbers = [1, 2, 3, 4]  # List is an iterable


for num in numbers:  # We can loop over it
    print(num)




---

2. What is an Iterator?

An iterator is an object that remembers where it is in an iterable and returns one item at a time when asked.

It implements both _iter() and __next_().

It does not store all values in memory at once (useful for large data).

It remembers its state between calls.


Example of an iterator:

numbers = iter([1, 2, 3, 4])  # Using iter() to create an iterator

print(next(numbers))  # Output: 1
print(next(numbers))  # Output: 2
print(next(numbers))  # Output: 3
print(next(numbers))  # Output: 4
# print(next(numbers))  # This will raise StopIteration




Q6. Explain the concept of generators in Python and how they are defined.

Ans. Generators in Python

Generators are a special type of iterator that allow for lazy evaluation—meaning they generate values one at a time as they are needed, instead of storing them all in memory.


---

1. How Are Generators Defined?

Generators are defined like normal functions but use the yield keyword instead of return.

Example: Simple Generator

def count_up_to(n):
    count = 1
    while count <= n:
        yield count  # Yield instead of return
        count += 1

# Using the generator
gen = count_up_to(5)
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # Output: 3

Unlike a normal function, the execution pauses when yield is encountered and resumes from the same place when next() is called.

Once all values are yielded, calling next() again will raise StopIteration.



---

2. How Generators Work

A generator function automatically creates an iterator. It doesn’t return all values at once but generates them on demand.

Key Features:

Uses yield instead of return.

Saves memory because it doesn't store all values



Q7. What are the advantages of using generators over regular functions?

Ans. Advantages of Using Generators Over Regular Functions

Generators provide several benefits compared to regular functions, especially when dealing with large datasets or complex computations.


---

1. Memory Efficiency (Lazy Evaluation)

Generators produce values on demand, instead of storing all values in memory at once.

This is especially useful for processing large files, streams, or infinite sequences.


Example:

def count_up_to(n):
    count = 1
    while count <= n:
        yield count  # Generates values one at a time
        count += 1

gen = count_up_to(1000000)  # Uses almost no memory

In contrast, a regular function:

def count_list(n):
    return list(range(1, n + 1))  # Stores all values in memory

nums = count_list(1000000)  # Uses a lot of memory

2. Faster Execution (No Precomputing)

Regular functions compute and return the entire result at once, while generators pause and resume execution using yield, making them faster in some cases.

Example:

def generate_squares(n):
    for i in range(n):
        yield i * i  # Computes only when needed

This approach avoids computing unnecessary values upfront.


---

3. Simplified Code (No Need for Custom Iterators)

Instead of implementing _iter() and __next_() manually in a class, generators provide an easier way to create iterators.

Using a generator:

def countdown(n):
    while n > 0:
        yield n
        n -= 1

Without a generator (manual iterator):

class Countdown:
    def _init_(self, n):
        self.n = n
    
    def _iter_(self):
        return self
    
    def _next_(self):
        if self.n <= 0:
            raise StopIteration
        self.n -= 1
        return self.n + 1

✅ Generator: Shorter and easier to read.
❌ Custom Iterator: Requires more code and handling.


---

4. Infinite Sequences (Generators Can Keep Running)

A generator can run indefinitely, generating values as needed, while a normal function must return all values at once.

Example: Infinite Fibonacci Sequence

def fibonacci():
    a, b = 0, 1
    while True:
        yield a  # Infinite sequence
        a, b = b, a + b

fib = fibonacci()
for _ in range(5):
    print(next(fib))  # Output: 0, 1, 1, 2, 3

A regular function would have to return an infinitely large list, which is impractical.


---

5. Better Performance for Large Data Processing

When working with large files, generators process data line by line instead of loading everything into memory.

Example: Reading a file line by line with a generator

def read_large_file(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.strip()  # Reads one line at a time

for line in read_large_file("bigfile.txt"):
    print(line)  # Processes one line at a time

Q9. Explain the purpose and usage of the map() function in Python.

Ans. The map() function is used to apply a function to each item of an iterable (e.g., list, tuple, set) and return an iterator with the modified values. It helps in writing concise and efficient code.


---

1. Syntax of map()

map(function, iterable)

function – A function that is applied to each item in the iterable.

iterable – A sequence (list, tuple, etc.) whose items are processed.

Returns a map object (iterator), which can be converted to a list or another collection.



---

2. Example: Using map() with a Regular Function

def square(x):
    return x * x

numbers = [1, 2, 3, 4]
result = map(square, numbers)
print(list(result))  # Output: [1, 4, 9, 16]

3. Example: Using map() with a Lambda Function

A lambda function makes map() even more concise:

numbers = [1, 2, 3, 4]
result = list(map(lambda x: x * x, numbers))
print(result)  # Output: [1, 4, 9, 16]

4. Using map() with Multiple Iterables

map() can take multiple iterables and apply the function to corresponding elements.

Example: Adding Two Lists Element-wise

a = [1, 2, 3]
b = [4, 5, 6]

result = list(map(lambda x, y: x + y, a, b))
print(result)  # Output: [5, 7, 9]


5. Using map() to Convert Data Types

Example: Converting a List of Strings to Integers

strings = ["1", "2", "3"]
numbers = list(map(int, strings))
print(numbers)  # Output: [1, 2, 3]

6. Using map() with Built-in Functions

Example: Converting Strings to Uppercase

words = ["hello", "world"]
uppercase_words = list(map(str.upper, words))
print(uppercase_words)  # Output: ['HELLO', 'WORLD']


7. map() vs. List Comprehension

Both map() and list comprehensions can transform data, but they have differences:

Example Comparison

# Using map()
result = list(map(lambda x: x * 2, [1, 2, 3]))
print(result)  # Output: [2, 4, 6]

# Using list comprehension
result = [x * 2 for x in [1, 2, 3]]
print(result)  # Output: [2, 4, 6]

```python
# Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.

# ans.

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Checking prime numbers from 1 to 200
primes = [num for num in range(1, 201) if is_prime(num)]
print(primes)
```

```python
# Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

# Ans.

def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent


for power in powers_of_two(5):
    print(power, end=" ")
```

```python
# Q7. Implement a generator function that reads a file line by line and yields each line as a string

# Ans.

def read_file_line_by_line(file_path):
    with open(file_path, "r", encoding="utf-8") as file:
        for line in file:
            yield line.strip()  # Removes leading/trailing whitespace

# Example usage
file_path = "example.txt"
for line in read_file_line_by_line(file_path):
    print(line)
```

```python
# Q10. Create a Python program that uses filter() to remove all the vowels from a given string.

# Ans.

# Function to check if a character is not a vowel
is_consonant = lambda char: char.lower() not in 'aeiou'

# Function to remove vowels from a string
def remove_vowels(string):
    return ''.join(filter(is_consonant, string))


input_string = "Hello, World!"
output_string = remove_vowels(input_string)

print(output_string)
```
